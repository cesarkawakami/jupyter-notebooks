```python
%pylab inline
from ipywidgets import interact
# import seaborn as sns
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
@interact(elev=(0, 90, 1), azim=(0, 360, 1))
def go(elev=45, azim=147):
# elev, azim = 45, 147
    u_s = array([u for u in linspace(0.0, 1.0, 100) if log1p(u) >= u**2.0])
    v_s = linspace(0.0, 1.0, 100)
    u = outer(u_s, ones_like(v_s))
    v = outer(ones_like(u_s), v_s)
    x = u
    y = u**2 + (log1p(u) - u**2) * v
    z = log1p(x) - x**2

    fig = figure(figsize=(8, 8))
    ax = fig.add_subplot(1, 1, 1, projection="3d")
    ax.plot_surface(x, y, z, alpha=.8)

    x = u
    z = (log1p(u) - u**2) * v
    y = log1p(u)
    ax.plot_surface(x, y, z, alpha=.8)

    x = u
    z = (log1p(u) - u**2) * v
    y = u**2
    ax.plot_surface(x, y, z, alpha=.8)


    ax.view_init(elev, azim)
    show(fig)
```
